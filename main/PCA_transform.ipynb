```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from matplotlib.pyplot import imread
import numpy as np
```

```python
img = imread("LC08_L1TP_149035_20130705_20200912_02_T1_refl.tif")
img = img.astype(np.uint8)
img = img/255
img = img.mean(axis=2)
Original_img=img
print(img.astype(float))
plt.imshow(img)
img.shape
```

```python
Matrices=[]

def transform(percentage):
  percentage = percentage/100
  img_pca = PCA(n_components=percentage).fit(img)
  transformed = img_pca.transform(img)
  projection = img_pca.inverse_transform(transformed)
  Matrices.append(projection)
  plt.imshow(projection)
```

```python
Matrices[0]
```

```python
transform(10)
```

```python
mse = []
for i in range(3):
  mse.append(((Original_img - Matrices[i])**2).mean(axis=None))
```
